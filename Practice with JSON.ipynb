```python
import json
import pandas as pd
with open(r"\Users\Preston Overson\Documents\Python Scripts\Data\airlines.JSON") as f:
    air_data = json.load(f)
```

```python
from pandas.io.json import json_normalize
air_df = json_normalize(air_data)
air_df.head(5)
```

```python
air_df.describe()
```

```python
air_df.info()
```

```python
atl_df =  air_df['airport.code']=='ATL'
```

```python
atl_df.head()
```

```python
atl_df = air_df[atl_df]
```

```python
atl_dl_df =  atl_df['carrier.code']=='DL'
atl_dl_df = atl_df[atl_dl_df]
atl_dl_df.info()
```

```python
import seaborn as sns
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
sns.set(style='ticks', palette='Set2')
# shows graphs in notebook
%matplotlib inline


x = atl_dl_df['time.label']
y = atl_dl_df['statistics.# of delays.weather']
fig = sns.lineplot(x, y)
fig.set(xlabel='Time', ylabel='# Delays Due to Weather')
plt.show()
```

```python
atl_dl_df_2008 = atl_dl_df['time.year'] == 2008
atl_dl_df_2008 = atl_dl_df[atl_dl_df_2008]
atl_dl_df_2008.info()
```

```python
#Plotting delas for delta in 2008

x = atl_dl_df_2008['time.month']
y = atl_dl_df_2008['statistics.minutes delayed.carrier']
fig = sns.lineplot(x, y)
fig.set(xlabel='Month', ylabel='Delta Delays (minutes)')
plt.show()
```
